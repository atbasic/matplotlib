```python
#x values
x = [1,2,3,4,5,6,7]

#y values

y = [2,14,6,28,10,32,14]

plt.plot(x,y,c="blue", linewidth='1', marker='x', ms=10, mfc= "g", mec="g") #plotting the points

fontxlabel = {'family' : 'serif','color': 'blue','size': 20}
fontylabel = {'family' : 'serif','color': 'darkred','size': 20}
plt.xlabel("x-axis",fontdict = fontxlabel)
plt.ylabel("y-axis",fontdict = fontylabel)
plt.tick_params(axis="x", labelsize=14, labelcolor="orange",labelrotation=-60)
plt.tick_params(axis="y",labelsize=10,labelcolor="red", labelrotation=-60)
plt.title("my first plot",loc="right")
plt.grid( axis="x",color="green")
plt.show() #function to show the plot
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x = np.array([10,50,60,80,12,13,90,0,85,53])
y = np.array([20,35,42,47,23,24,49,10,47,36])
plt.scatter(x,y, color="red", s=25)
fontxlabel = {'family' : 'serif','color': 'blue','size': 20}
fontylabel = {'family' : 'serif','color': 'darkred','size': 20}
plt.xlabel("attendance",fontdict = fontxlabel)
plt.ylabel("marks",fontdict = fontylabel)
plt.show
```

```python
x = np.array([10,50,60,80,12,13,90,0,85,53])
y = np.array([20,35,42,47,23,24,49,10,47,36])
colors = np.array(['red','green','orange','blue','yellow','pink','black','cyan','darkred','magenta'])

plt.scatter(x,y, color=colors, s=25)
fontxlabel = {'family' : 'serif','color': 'blue','size': 20}
fontylabel = {'family' : 'serif','color': 'darkred','size': 20}
plt.xlabel("attendance",fontdict = fontxlabel)
plt.ylabel("marks",fontdict = fontylabel)
plt.show
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x= np.linspace(0,10,1000)
plt.plot(x,np.sin(x), '--b', label='sin')
plt.plot(x,np.cos(x), 'r',label='cos')
plt.legend(loc='lower left')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x= np.linspace(0,20,1000)
plt.plot(x,np.cos(x), 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
x= np.linspace(0,20,1000)
plt.plot(x,np.cos(x), 'r')
plt.axis('equal')
plt.annotate('local maximum',xy=(6.5,1),xytext=(10,4), arrowprops=dict(facecolor='black',shrink=0.05))
plt.annotate('local minimum',xy=(15,-1),xytext=(2,-6), arrowprops=dict(facecolor='pink',shrink=0.05))
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np
from mpl_toolkits import mplot3d
%matplotlib notebook
ax=plt.axes(projection = '3d')
z = np.linspace(0,1,100)
x = z*np.sin(25*z)
y = z*np.cos(25*z)
ax.plot3D(x,y,z, 'red')
ax.set_title('3d plot')
plt.show()
```
